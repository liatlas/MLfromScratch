```python
def calc_variance(x: pd.Series) -> np.float64:
    '''
    Returns the variance of a series
    '''
    sum: np.float64 = 0.0

    mu_x: np.float64 = np.mean(x)
    n: int = x.shape[0]

    for i in range(0, n):
        sum += (x[i] - mu_x) ** 2

    return sum / (n-1)
```

```python
def calc_covariance(x: pd.Series, y: pd.Series) -> np.float64:
    '''
    Return cov(a,b) given two series of equal length
    '''
    
    sum = 0
    mu_x = np.mean(x)
    mu_y = np.mean(y)
    n = x.shape[0]

    for i in range(0, n):
        sum += (x[i] - mu_x) * (y[i] - mu_y)

    sum /= length - 1

    return sum
```
